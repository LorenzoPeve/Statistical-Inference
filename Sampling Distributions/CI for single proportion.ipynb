## Confidence Interval for Single Proportion

```python
import pandas as pd
import seaborn as sns
import os
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import random
```

#### Original Sample

$ N = 250 \\
 \hat p = 180/100$

```python
SIMULATIONS = 5000
N = 250 # Original sample
p_hat = 180/N # Count in original sample
```

Helper function

```python
def add_text(ax, t, xy):
    ax = ax.annotate(t, xy=xy, xycoords='data',
        xytext=(30, 0), textcoords='offset points')
```

```python
p = []
for simulation in range(SIMULATIONS):
    p.append(sum(random.choices([1,0], weights=[p_hat, 1-p_hat], k = N))/N)
    
avg = sum(p) / len(p)
```

$ SE_{\hat p} = \sqrt{\dfrac{\hat p (1 - \hat p}{n}}$

```python
SE = np.sqrt( (p_hat * (1-p_hat))/ (N))
SE
```

```python
bins = np.arange(p_hat - 3 * SE, p_hat + 3 * SE, 0.02)
if len(bins) < 15:
    bins = 15

fig, ax = plt.subplots(figsize =(8,6))
ax = sns.histplot(x = p, bins = bins, ax = ax)
ax.axvline(x=avg, linewidth=3, color='firebrick', linestyle = 'dashed')
add_text(ax, f'Mean: {avg:.4f}, SE = {np.std(p, ddof = 1):.4f}', (avg+0.01, 1000))
```

```python
low = p_hat - 2*SE
high = p_hat + 2*SE
CI = [low, high]
CI
```
