## Confidence Interval for Single Mean

```python
import pandas as pd
import seaborn as sns
import os
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import random
```

#### Original Sample:  Number of ants on a sandwich

$ N = 8 $

```python
sample = [43, 59, 22, 25, 36, 47, 19, 21]
```

```python
SIMULATIONS = 5000
```

Helper function

```python
def add_text(ax, t, xy):
    ax = ax.annotate(t, xy=xy, xycoords='data',
        xytext=(30, 0), textcoords='offset points')
```

```python
p = []
N = len(sample)
for simulation in range(SIMULATIONS):
    p.append(sum(random.choices(sample, k = N))/N)
    
avg = sum(p) / len(p)
```

```python
fig, ax = plt.subplots(figsize =(8,6))
ax = sns.histplot(x = p, bins = 15, ax = ax)
ax.axvline(x=avg, linewidth=3, color='firebrick', linestyle = 'dashed')
add_text(ax, f'Mean: {avg:.4f}, SE = {np.std(p, ddof = 1):.4f}', (avg+0.01, 800))
```

```python
SE = 4.8752
x_bar = np.mean(sample)
low = x_bar - 2*SE
high = x_bar + 2*SE
CI = [low, high]
CI
```
